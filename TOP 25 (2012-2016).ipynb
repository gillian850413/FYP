```python
# import numpy as np 
import pandas as pd

data_folder = r"C:\Users\pipi3\Desktop\Course\year3_sem3 (Japan)\FYP\input_data";
df = pd.read_csv(data_folder+"\\Top 25(2012-2016).csv")
df = df.fillna(0)
df = df.sort_values(by='Total', ascending=0)
df
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from pandas.plotting import table
```

```python
df = df[:15]

f, ax1 = plt.subplots(1, figsize=(20,10))

# Set the bar width
bar_width = 0.5

# positions of the left bar-boundaries
bar_l = [i+1 for i in range(len(df['2012']))]

# positions of the x-axis ticks (center of the bars as bar labels)
tick_pos = [i+(bar_width/2) for i in bar_l]

# Create a bar plot, in position bar_1
ax1.bar(bar_l,
        df['2012'],# using the 2012 data      
        width=bar_width,# set the width
        label='2012',# with the label 2012
        alpha=0.5,# with alpha 0.5
        color='#F4561D') # with color

ax1.bar(bar_l,
        df['2013'],
        width=bar_width,
        bottom=df['2012'],# with 2012 on the bottom
        label='2013',
        alpha=0.5,
        color='#F1911E')

ax1.bar(bar_l,
        df['2014'],
        width=bar_width,
        bottom=[i+j for i,j in zip(df['2012'],df['2013'])],
        label='2014',
        alpha=0.5,
        color='#F1BD1A')

ax1.bar(bar_l,
        df['2015'],
        width=bar_width,
        bottom=[i+j+k for i,j,k in zip(df['2012'],df['2013'],df['2014'])],
        label='2015',
        alpha=0.5,
        color='#000000')

ax1.bar(bar_l,
        df['2016'],
        width=bar_width,
        bottom=[i+j+k+l for i,j,k,l in zip(df['2012'],df['2013'],df['2014'],df['2015'])],
        label='2016',
        alpha=0.5,
        color='#FFF000')

# set the x ticks with names
plt.xticks(tick_pos, df['Company'])
plt.title('Top 15 Companies Game Revenue')

# Set the label and legends
ax1.set_ylabel("Total Revenue(in Million US)")
ax1.set_xlabel("Company")
plt.legend(loc='upper right')

# Set a buffer around the edge
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
```

```python
df = df[:10]

n_groups = 10
fig, ax = plt.subplots(1, figsize=(40,12))
index = np.arange(n_groups)
bar_width = 0.15
opacity = 0.4
error_config = {'ecolor': '0.3'}


com1 = plt.bar(index , df['2012'], bar_width,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='2012')

com2 = plt.bar(index + bar_width*1, df['2013'], bar_width,
                 alpha=opacity,
                 color='yellow',
                 error_kw=error_config,
                 label='2013')


com3 = plt.bar(index + bar_width*2, df['2014'], bar_width,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='2014')

com4 = plt.bar(index + bar_width*3, df['2015'], bar_width,
                 alpha=opacity,
                 color='green',
                 error_kw=error_config,
                 label='2015')

com5 = plt.bar(index + bar_width*4, df['2016'], bar_width,
                 alpha=opacity,
                 color='black',
                 error_kw=error_config,
                 label='2016')


plt.xlabel('Company')
plt.ylabel('Revenue (in Million US)')
plt.title('Top 10 Companies Game Revenue')
plt.xticks(index + bar_width*2 / 2, (df['Company'][:10]))
plt.legend(loc='best')

ax2 = ax.twinx()
ax2.plot(ax.get_xticks(),df['Total'].values, linestyle='-', marker='o', linewidth=2.0, label = 'Total')
ax2.set_ylabel('Total Revenue (in Million US)')
ax2.legend(loc='best')


plt.tight_layout()
plt.show()
```
